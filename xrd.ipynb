# Tâche 7: Diffactogramme

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer

# Clé API MP
with MPRester("1ziZTLbVkqHKwpma ") as m:
    # Structure du matériau mp-7575
    structure = m.get_structure_by_material_id("mp-7575")

struct_c = SpacegroupAnalyzer(structure).get_conventional_standard_structure()#structure conventionelle

#XRDCalculator avec rayons x 'CuKa'
xrd = XRDCalculator('CuKa')
#Calcul du modèle de diffraction
pattern = xrd.get_pattern(struct_c)
```

## Trois premiers pics du diffractogramme de LiZnN par rayons X de λ=1.54060 Å

```python

for i in range(3):
    two_theta = pattern.x[i]
    I = pattern.y[i]
    hkl = pattern.hkls[i]
    print("\nPic ", i+1)
    print("Angle 2theta(degrés) = ", two_theta)
    print("Intensité(%) = ", I)
    print("Indices de Miller = ", hkl)
```

2theta est l'angle entre le rayon incident et diffracté, theta(= 2theta/2) est l'angle entre le rayon (incident ou diffracté) et le plan réticulaire mis en conditions de diffraction et hkl sont les indices de Miller du plan concerné.

```python
xrd.show_plot(struct_c, two_theta_range=(30,60),annotate_peaks=True,ax=None,with_labels=True)
```
